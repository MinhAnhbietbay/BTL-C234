```python
import pandas as pd
```

```python
df = pd.read_csv("../dataset/bonbanh_cleandata.xls")
df.head(5)
```

```python
df.dtypes
```

```python
vis_cols_to_scale = ['price_num', 'mileage']
vis_scaling_params = {}
visualization_df = df.copy()

for col in vis_cols_to_scale:
    mean = visualization_df[col].mean()               # trung bình
    std = visualization_df[col].std()                 # độ lệch chuẩn
    visualization_df[col] = (visualization_df[col] - mean) / std   # Z-score normalization

    vis_scaling_params[col] = {'mean': float(mean), 'std': float(std)}
vis_scaling_params
```

```python
visualization_df.to_csv("../dataset/visualization_data.csv", index=False)
```

```python
numeric_cols_model = ['year', 'price_num', 'mileage', 'seats', 'engine_capacity']

df_model = df.copy()
scaling_params_full = {}
```

```python
for col in numeric_cols_model:
    mean = df_model[col].mean()
    std = df_model[col].std()
    df_model[col] = (df_model[col] - mean) / std
    scaling_params_full[col] = {'mean': float(mean), 'std': float(std)}
scaling_params_full
```

```python
# encoder car_type, color, engine, transmission

cat_cols = ['car_type', 'color', 'transmission', 'fuel_type']
cols = numeric_cols_model + cat_cols

df_clean = df_model[cols].copy()

df_final = pd.get_dummies(df_clean, columns=cat_cols, dummy_na=False)
dummy_cols = [col for col in df_final.columns if col not in numeric_cols_model]
dummy_cols
```

```python
df_final.head()
```

```python
df_final.to_csv('../dataset/normalization_One-hot-Encoder.csv', index=False)
pd.DataFrame.from_dict(scaling_params_full, orient='index').to_csv("../dataset/scaling_params_full.csv")
```
